This notebook queries for all SGCN species that are listed with FWS as candidate species. It shows some basic information from the Taxa Information Registry and lists the states where the species has been listed.

```python
import requests
from IPython.display import display
import json
```

```python
r_listingStatus = requests.get("https://gc2.datadistillery.org/api/v1/sql/bcb?q=SELECT registration,tess FROM tir.tir WHERE tess->>'STATUS' = 'C' AND registration->>'source' = 'SGCN'").json()

for record in r_listingStatus["features"]:
    _registration = json.loads(record["properties"]["registration"])
    _registeredName = _registration["scientificname"]
    _tess = json.loads(record["properties"]["tess"])
    r_states = requests.get("https://gc2.datadistillery.org/api/v1/sql/bcb?q=SELECT DISTINCT sgcn_state FROM sgcn.sgcn WHERE scientificname_submitted = '"+_registeredName+"' ORDER BY sgcn_state").json()
    
    print ("Candidate Species Scientific Name: ", _registeredName)
    print ("FWS Common Name: ", _tess["COMNAME"])
    print ("Lead Region: ", _tess["LEAD_REGION"])
    for listingStatus in _tess["listingStatus"]:
        print ("Population Description: ", listingStatus["POP_DESC"], "(", listingStatus["STATUS"], ")")
    print ("FWS Entity ID: ", _tess["ENTITY_ID"])
    print ("Date Last Checked: ", _tess["dateCached"])

    states = []
    for sgcnRecord in r_states["features"]:
        states.append(sgcnRecord["properties"]["sgcn_state"])
    print ("States with species listed (2005-2015): ", states)
    print ("------------")
    
    
```
